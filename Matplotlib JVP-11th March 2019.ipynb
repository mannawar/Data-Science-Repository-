```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
fig = plt.figure()
ax = plt.axes()
```

```python
x = np.linspace (0, 10, 1000)
ax.plot(x, np.sin(x));
```

```python
plt.plot(x, np.sin(x));
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x));
```

```python
plt.plot(x, np.sin(x-0), color = 'blue', linestyle ='dashed')
plt.plot(x, np.sin(x-1), color ='green', linestyle ='dotted')
plt.plot(x, np.sin(x-2), color ='0.75', linestyle = 'dashdot')
plt.plot(x, np.sin(x-3), color ='#FFDD44', linestyle = 'solid')
```

```python
plt.plot(x, x+0, '-g')
plt.plot(x, x+1, '--c')
plt.plot(x, x+2, '-.k')
plt.plot(x, x+3, ':r')
```

```python
plt.plot(x, np.cos(x))
plt.xlim(-2,11)
plt.ylim(-2, 2)
```

```python
plt.plot(x, np.sin(x))
plt.axis([0, 10, -1, 1])
```

```python
plt.plot(x, np.sin(x))
plt.axis([0, 10, -1, 1], 'tight', 'scaled', 'equal')
```

```python
# Labelling plots
plt.plot(x, np.sin(x))
plt.title("A sine curve")
```

```python
plt.plot(x, np.sin(x), '-g', label = 'sin(x)')
plt.plot(x, np.cos(x), ':', label = 'cos(x)')
plt.axis('equal')
plt.legend();
```

```python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim= (-2, 2),
      xlabel = 'x', ylabel ='sin(x)',
      title = 'A simple plot');
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
plt.plot(x, y, 'o', color='black')
```

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
        label = "marker='{0}'".format(marker))
plt.legend(numpoints = 1)
plt.xlim(0,1.8);
```

```python
plt.plot(x, y, '-ok');
```

```python
plt.plot(x, y, '-p', color = 'green', markersize = 15, linewidth= 4)
```

```python
plt.ylim(-1.2, 1.2);
```

```python
plt.scatter(x, y, marker = '>')
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000*rng.rand(100)
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap= 'viridis')
plt.colorbar();
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
features =iris.data.T
plt.scatter(features[0], features[1], aplha = 0.2,
           s=100*features[3], c= iris.target, cmap= 'viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
```
