```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(iris.data, columns =iris.feature_names)
df
```

```python
df.head()
```

```python
df['target']= iris.target
```

```python
df.head()
```

```python
iris.target_names
```

```python
df[df.target==2].head()
```

```python
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
df
```

```python
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
df0 = df[df.target ==0]
df1 = df[df.target ==1]
df2 = df[df.target ==2]
```

```python
df0.head()
```

```python
plt.xlabel('sepal length (cm)')
plt.ylabel ('sepal width (cm)')
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color ='green', marker ='+')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color ='blue', marker ='x')
```

```python
plt.xlabel('petal length (cm)')
plt.ylabel ('petal width (cm)')
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'], color ='green', marker ='+')
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'], color ='blue', marker ='x')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(['target', 'flower_name'], axis = 'columns')
x.head()
```

```python
y = df['target']
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.svm import SVC
model = SVC (kernel = 'linear')
```

```python
model.fit (x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
from sklearn.datasets import load_digits
```

```python
dir (digits)
```

```python
df1= pd.DataFrame(digits.data)
df1.head()
```

```python
df1['target'] = digits.target
```

```python
df1
```

```python
digits.target_names
```

```python
x1 = df1.drop(['target'], axis = 'columns')
x1.head()
```

```python
x1_train, x1_test, y1_train, y1_test = train_test_split(x1, y1, test_size=0.2)
```

```python
len(x1_train)
```

```python
len(x1_test)
```

```python
from sklearn.svm import SVC
model = SVC (kernel = 'poly', gamma = 'auto')
```

```python
model.fit (x1_train, y1_train)
```

```python
model.score(x1_test, y1_test)
```
