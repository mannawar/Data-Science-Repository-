```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_excel("c:\\Users\\Mannawar Hussain\\Desktop\\Data\\Meter reading data.xlsx")
```

```python
df    # after extracting data from excel as we can see there are so many na values, so i will apply missing value treatment in which all na will be replaced with 'o'
```

```python
df.head()
```

```python
df = df.fillna(value = 0, axis = 1)
df
```

```python
# Representation of consumption of data for year 2018 and 2019 
%matplotlib inline
plt.xlabel ('consumption2018')
plt.ylabel ('consumption2019')
plt.scatter(df.consumption2018, df.consumption2019, color = 'red', marker= '+')
```

```python
# Applying linear regression technique to compare data in year 2018 and 2019
reg = linear_model.LinearRegression()
reg.fit(df[['consumption2018']], df.consumption2019)
```

```python
reg.predict ([[16]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
%matplotlib inline
plt.xlabel ('consumption2018', fontsize =20)
plt.ylabel ('consumption2019', fontsize =20)
plt.scatter(df.consumption2018, df.consumption2019, color = 'red', marker= '+')
plt.plot(df.consumption2019,reg.predict(df[['consumption2019']]), color='blue')
```
