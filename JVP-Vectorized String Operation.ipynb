```python
data = pd.read_csv("c:\\Users\\Mannawar Hussain\\Desktop\\Fremont_Bridge_Hourly_Bicycle_Counts_by_Month_October_2012_to_present.csv", index_col='Date', parse_dates = True)
```

```python
data.head()
```

```python
data.columns =([['West', 'East']])
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn; seaborn.set()
```

```python
data.plot()
plt.ylabel('Hourly Bicycle Count');
```

```python
weekly = data.resample('W').sum()
weekly.plot(style =[ ':', '--', '-'])
plt.ylabel('weekly bicycle count');
```

```python
daily= data.resample ('D').sum()
daily.rolling(30, center = True).sum().plot(style =[':', '--', '-'])
plt.ylabel('mean hourly count');
```

```python
daily.rolling(50, center = True, win_type = 'boxcar').sum(std = 10).plot(style =[':', '--', '-'])
```

```python
daily.rolling(50, center = True, win_type = 'gaussian').sum(std = 10).plot(style =[':', '--', '-'])
```

```python
daily.rolling(50, center = True, win_type = 'triang').sum(std = 10).plot(style =[':', '--', '-'])
```

```python
daily.rolling(50, center = True, win_type = 'kaiser').sum(std = 10, beta =.1 ).plot(style =[':', '--', '-'])
```

```python
daily.rolling(50, center = True, win_type = 'slepian').sum(width =100).plot(style =[':', '--', '-'])
```

```python
by_time =data.groupby(data.index.time).mean()
hourly_ticks = 4 * 60 * 60 * np.arange(6)
by_time.plot(xticks =hourly_ticks, style = [':', '--', '-']);
```

```python
by_weekday = data.groupby(data.index.dayofweek).mean()
by_weekday.index = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
by_weekday.plot(style= [':', '--', '-']);
```

```python
weekend =np.where(data.index.weekday < 5, 'weekday', 'weekend')
by_time = data.groupby([weekend, data.index.time]).mean()
```

```python
fig, ax =plt.subplots(1, 2, figsize=(14, 5))
by_time.loc['weekday'].plot(ax = ax[0], title = 'weekdays',
                          xticks = hourly_ticks, style =[':', '--', '-'])
by_time.loc['weekend'].plot(ax = ax[1], title = 'weekends',
                          xticks= hourly_ticks, style = [':', '--','-'])
                      
```

```python
fig, ax =plt.subplots(1, 2, figsize=(14, 5))
by_time.ix['weekday'].plot(ax = ax[0], title = 'weekdays',
                          xticks = hourly_ticks, style =[':', '--', '-'])
by_time.ix['weekend'].plot(ax = ax[1], title = 'weekends',
                          xticks= hourly_ticks, style = [':', '--','-'])
                      
```
